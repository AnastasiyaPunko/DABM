```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
import os

if not os.path.exists("images"):
    os.mkdir("images")
```

```python
bacteria = pd.read_csv('/content/Protein lengths - Бактерии.csv')
eukaryotes = pd.read_csv('/content/Protein lengths - Эукариоты.csv')

df = bacteria.drop(columns = ['Name','Length','Name.1','Length.1','Name.2','Length.2','Name.3','Length.3','Name.4','Name.5','Length.5'])
dff = eukaryotes.drop(columns = ['Name','Length','Name.1','Length.1','Name.2','Length.2','Name.3','Length.3','Name.4','Length.4','Name.5'])
```

```python
df = df.dropna()
dff = dff.dropna()
print(df.shape)
print(dff.shape)
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(
    x=df['Length.4'],
    #histnorm='percent',
    name='Mycobacterium tuberculosis', # name used in legend and hover labels
    xbins=dict( # bins used for histogram
        size=55
    ),
    marker_color='darkgreen',
    opacity=1
))
fig.add_trace(go.Histogram(
    x=dff['Length.5'],
    #histnorm='percent',
    name='Saccharomyces cerevisiae',
    xbins=dict(
        size=55
    ),
    marker_color='#330C73',
    opacity=1
))

fig.update_layout(
    xaxis_title_text='Длины белков M.tuberculosis и S.cerevisiae', # xaxis label
    bargap=0.2, # gap between bars of adjacent location coordinates
    bargroupgap=0.1 # gap between bars of the same location coordinates
)

fig.show()
```

```python
fig.write_image("images/hist1.jpeg")
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(
    x=df['Length.4'],
    #histnorm='percent',
    name='Mycobacterium tuberculosis', # name used in legend and hover labels
    xbins=dict( # bins used for histogram
        size=225
    ),
    marker_color='darkgreen',
    opacity=1
))
fig.add_trace(go.Histogram(
    x=dff['Length.5'],
    #histnorm='percent',
    name='Saccharomyces cerevisiae',
    xbins=dict(
        size=225
    ),
    marker_color='#330C73',
    opacity=1
))

fig.update_layout(
    xaxis_title_text='Длины белков M.tuberculosis и S.cerevisiae', # xaxis label
    bargap=0.2, # gap between bars of adjacent location coordinates
    bargroupgap=0.1 # gap between bars of the same location coordinates
)

fig.show()
```

```python
fig.write_image("images/hist2.jpeg")
```

```python
np.random.seed(1)

fig = go.Figure()
fig.add_trace(go.Box(y=df['Length.4'],
    name='Mycobacterium tuberculosis',
    marker_color='darkgreen'))
fig.add_trace(go.Box(y=dff['Length.5'],
    name='Saccharomyces cerevisiae',
    marker_color='#330C73'))

fig.show()
```

```python
fig.write_image("images/box.jpeg")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.kdeplot(df['Length.4'],
            color='darkgreen', label='Mycobacterium tuberculosis', cumulative=True, ax=ax)
sns.kdeplot(dff['Length.5'],
            color='#330C73', label='Saccharomyces cerevisiae', cumulative=True, ax=ax)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.histplot(df['Length.4'],
            color='darkgreen', label='Mycobacterium tuberculosis', element="step", fill=False,cumulative=True, ax=ax, stat="density", bins=len(df))
sns.histplot(dff['Length.5'],
            color='#330C73', label='Saccharomyces cerevisiae',element="step", fill=False, cumulative=True, ax=ax, stat="density", bins=len(dff))
ax.legend()
plt.tight_layout()
plt.show()
```
